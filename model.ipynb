```
import pandas as pd
import numpy as np
```

```
credits_df = pd.read_csv(path+'/tmdb_5000_credits.csv')
movies_df=pd.read_csv(path+'/tmdb_5000_movies.csv')
```

```
import ast
```

```
movies_df.drop(columns=['budget','homepage','original_language','original_title','popularity','production_countries','release_date','revenue','runtime','spoken_languages','status','tagline','vote_average','vote_count'], inplace=True)
```

```
credits_df.head(5)
credits_df.rename(columns={'movie_id': 'id'},  inplace=True)
credits_df.head(5)
```

```
main_df = pd.merge(credits_df, movies_df, on='id')
```

```
main_df.isnull().sum()
main_df.dropna(how='any', inplace=True)
main_df.isnull().sum()
```

```
main_df.drop(columns=['title_x'], inplace=True)
main_df.rename(columns={'title_y': 'title'}, inplace=True)
```

```
main_df.head(5)
```

```
for i in ['genres','keywords','production_companies','crew','cast']:
    main_df[i]=main_df[i].apply(ast.literal_eval)
```

```
main_df['cast'][0]
```

```
def transform(text):
    y=[]
    z=0
    for i in text :
        if z !=3:
            y.append(text[z]['name'])
            z +=1
        else:
            break
    return y
```

```
main_df['cast']=main_df['cast'].apply(transform)
```

```
main_df.head(5)
```

```
main_df['crew'][0]
```

```
def crew(text):
    y=[]
    for i in text :
        if i['job']=='Director':
            y.append(i['name'])

    return y
```

```
main_df['crew']=main_df['crew'].apply(crew)
```

```
main_df.head(5)
```

```
main_df['genres'][0]
```

```
def genres_keywords(text):
    y=[]
    for i in text :
            y.append(i['name'])

    return y
```

```
main_df['genres']=main_df['genres'].apply(genres_keywords)
```

```
main_df.head(5)
```

```
main_df['keywords']=main_df['keywords'].apply(genres_keywords)
```

```
main_df.head(5)
```

```
main_df['production_companies'][0]
```

```
main_df['production_companies']=main_df['production_companies'].apply(genres_keywords)
```

```
main_df.head(5)
```

```
def final_process(text):
    y=[]
    for i in text :

        y.append(i.replace(' ',''))
    return y
```

```
main_df['cast']=main_df['cast'].apply(final_process)
main_df['crew']=main_df['crew'].apply(final_process)
main_df['genres']=main_df['genres'].apply(final_process)
main_df['keywords']=main_df['keywords'].apply(final_process)
main_df['production_companies']=main_df['production_companies'].apply(final_process)
main_df.head(5)
```

```
def x(text):
    y=[]
    for i in text :
        y.append(i)
    return ' '.join(y)
```

```
main_df['cast']=main_df['cast'].apply(x)
main_df['crew']=main_df['crew'].apply(x)
main_df['genres']=main_df['genres'].apply(x)
main_df['keywords']=main_df['keywords'].apply(x)
main_df['production_companies']=main_df['production_companies'].apply(x)
```

```
main_df.head(5)
```

```
main_df['tags']=main_df['cast']+main_df['crew']+main_df['genres']+main_df['keywords']+main_df['production_companies']
```

```
main_df
```

```
final_df=main_df[['id','title','tags']]
```

```
final_df.reset_index(inplace=True)
```

```
import nltk
import string
```

```
def text_preprocessing(text):
    text=text.lower()
    text=nltk.word_tokenize(text)
    y=[]
    for i in text:
        if i.isalnum():
            y.append(i)
    text=y[:]
    y.clear()

    for i in text:
            if i  not in nltk.corpus.stopwords.words('english') and i not in string.punctuation:
                y.append(i)
    text=y[:]
    y.clear()
    for i in text:
        y.append(nltk.stem.porter.PorterStemmer().stem(i))

    return " ".join(y)
```

```
final_df['tags']=final_df['tags'].apply(text_preprocessing)
```

```
final_df.sample(5)
```

```
from sklearn.feature_extraction.text import TfidfVectorizer
```

```
tfidf=TfidfVectorizer()
```

```
vec=tfidf.fit_transform(final_df['tags']).toarray()
```

```
from sklearn.metrics.pairwise import cosine_similarity
sim=cosine_similarity(vec)
```

```
def recommendation(name):
    ind=final_df[final_df['title']==name].index[0]
    simmaliraty=sim[ind]
    movie_list=sorted(list(enumerate(simmaliraty)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(final_df.iloc[i[0]].title)
```

```
recommendation('Lage Raho Munna Bhai')
```
